## Exploratory Data Analysis ##

```python
#Sales Performance
total_sales = df_superstore['Sales'].sum()
total_profit = df_superstore['Profit'].sum()
print(f"Total Sales: {total_sales}, Total Profit: {total_profit}")
```

```python
#Sales by category

sales_by_category = df_superstore.groupby('Category')['Sales'].sum()
sales_by_category.plot(kind='bar', title='Sales by Category')
plt.show()
```

```python
#Sales by sub category
sales_by_subcategory = df_superstore.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
sales_by_subcategory.plot(kind='bar', title='Sales by Sub-Category')
plt.show()
```

```python
#Sales by region
sales_by_region = df_superstore.groupby('Region')['Sales'].sum()
sales_by_region.plot(kind='bar', title='Sales by Region')
plt.show()
```

```python
#Profit by region

profit_by_region = df_superstore.groupby('Region')['Profit'].sum()
profit_by_region.plot(kind='bar', title='Profit by Region')
plt.show()
```

```python
#Sales by states

sales_by_state = df_superstore.groupby('State')['Sales'].sum().sort_values(ascending=False)
sales_by_state.plot(kind='bar', title='Sales by State',width=0.8)
plt.figure(figsize=(15,6))
plt.show()
```

```python
#customer segment
segment_analysis = df_superstore.groupby('Segment')[['Sales', 'Profit']].sum()
segment_analysis.plot(kind='bar', title='Sales and Profit by Segment')
plt.show()
```
